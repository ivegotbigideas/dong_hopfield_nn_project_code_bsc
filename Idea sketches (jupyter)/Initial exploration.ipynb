```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

# Define number of neurons

```python
number_of_neurons = 1
```

# Define function to get V

```python
def g(u):
    return 1/(1+math.e**(-u))
```

# Define dui_dt function

```python
def dui_dt(u,a=2,T=1,I=0):
    sum = 0
    for j in range(number_of_neurons):
        sum += T*g(u)
    
    derivative = (-1/a)*u+(1/a)*(sum+I)
    return [u, derivative]
```

# Define stuff needed for plotting

```python
plot_size = 100
```

HOR and VER are the horizontal and vertical positions of the vectors

```python
hor = np.linspace(-plot_size,plot_size,20)
ver = np.linspace(-plot_size,plot_size,20)
HOR, VER = np.meshgrid(hor, ver)
```

UP and RIGHT are the directions of the vectors

```python
UP, RIGHT = np.zeros(HOR.shape), np.zeros(VER.shape)
```

```python
NI, NJ = HOR.shape
for i in range(NI):
    for j in range(NJ):
        u = HOR[i,j]
        derivative = dui_dt(u=u)
        UP[i,j] = derivative[0]
        RIGHT[i,j] = derivative[1]
```

# Plot the phase diagram

```python
plt.xlabel('$u$')
plt.ylabel('$du/dt$')
plt.xlim([-plot_size, plot_size])
plt.ylim([-plot_size, plot_size])
Q = plt.quiver(HOR, VER, UP, RIGHT)
```
