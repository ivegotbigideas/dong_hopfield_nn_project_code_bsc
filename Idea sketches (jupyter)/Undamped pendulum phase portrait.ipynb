```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(Y, t):
    y1, y2 = Y
    return [y2, -np.sin(y1)]
```

```python
y1 = np.linspace(-2.0, 8.0, 20)
y2 = np.linspace(-2.0, 2.0, 20)
```

Y1 is the horizonal position of the vector
Y2 is the vertical position of the vector

```python
Y1, Y2 = np.meshgrid(y1, y2)
```

```python
t = 0
```

```python
u, v = np.zeros(Y1.shape), np.zeros(Y2.shape)
```

```python
NI, NJ = Y1.shape
for i in range(NI):
    for j in range(NJ):
        x = Y1[i, j]
        y = Y2[i, j]
        yprime = f([x, y], t)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]
```

Y1, Y2 is the position of the arrow.

u, v is the direction of the arrow. u is how "right" it is. v is how "up" it is.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.quiver.html

```python
plt.xlabel('$y_1=y$')
plt.ylabel('$y_2=y\'$')
plt.xlim([-2, 8])
plt.ylim([-4, 4])
Q = plt.quiver(Y1, Y2, u, v)
```
